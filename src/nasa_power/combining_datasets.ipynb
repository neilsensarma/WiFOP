```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
def read_csvs(path, file1, file2):
    f1 = f'{path}/{file1}'
    f2 = f'{path}/{file2}'
    return pd.read_csv(f1), pd.read_csv(f2)
```

```python
common_path = '/Users/neil/Desktop/GaTech.nosync/Fall2025/CSE 6740/WiFOP/data/raw/power_ca_jul2023/'
```

```python
prec_df1, prec_df2 = read_csvs(common_path, 'PRECTOT_tile1.csv', 'PRECTOT_tile2.csv')
ps_df1, ps_df2 = read_csvs(common_path, 'PS_tile1.csv', 'PS_tile2.csv')
rh_df1, rh_df2 = read_csvs(common_path, 'RH2M_tile1.csv', 'RH2M_tile2.csv')
t2m_df1, t2m_df2 = read_csvs(common_path, 'T2M_tile1.csv', 'T2M_tile2.csv')
ws_df1, ws_df2 = read_csvs(common_path, 'WS10M_tile1.csv', 'WS10M_tile2.csv')
```

```python
dfs1 = [prec_df1, ps_df1, rh_df1, t2m_df1, ws_df1]
tile1_merged = reduce(lambda left, right: pd.merge(left, right, on=['LAT', 'LON', 'YEAR', 'DOY'], how='inner'), dfs1)
```

```python
tile1_merged
```

```python
dfs2 = [prec_df2, ps_df2, rh_df2, t2m_df2, ws_df2]
tile2_merged = reduce(lambda left, right: pd.merge(left, right, on=['LAT', 'LON', 'YEAR', 'DOY'], how='inner'), dfs2)
```

```python
tile2_merged
```

```python
merged_df = pd.concat([tile1_merged, tile2_merged], axis=0, ignore_index=True)
```

```python
merged_df.to_csv('NASA_POWER.csv', index=False)
```
